```python
# new_df = new_df.drop(['id'],axis=1)
import numpy as np
dfNoNas['nweek'] = dfNoNas['year_week'].map(lambda x: str(x)[-2:])
dfNoNas['nweek'] = dfNoNas['nweek'].astype(np.int64)
dfNoNas['date'] = pd.to_datetime(dfNoNas['date'])
dfNoNas['year'], dfNoNas['month'],dfNoNas['day'] = dfNoNas['date'].dt.year, dfNoNas['date'].dt.month, dfNoNas['date'].dt.day
dfNoNas = dfNoNas.drop(['date'],axis=1)
dfNoNas
```

```python
dfNoNas = dfNoNas.drop(['id'],axis=1)
dfNoNas
```

```python
dfNoNas = dfNoNas.drop(columns=['sales_units','year','month','year_week'])
```

```python
X = dfNoNas.drop(columns=['inventory_units','setmana'])
Y = dfNoNas['inventory_units']
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test   = train_test_split(X,Y,shuffle=True,random_state=69)
```

```python
import pickle
with open('best_params_xgboost_2.pkl','rb') as file2, open('best_params_xgboost.pkl','rb') as file1:
    best_params1 = pickle.load(file1)
    best_params2 = pickle.load(file2)
best_params1
```

```python
from sklearn.ensemble import GradientBoostingRegressor

gradientBoosting = GradientBoostingRegressor(**best_params2)

result_gradientBoosting = gradientBoosting.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error
y_predicted_test = gradientBoosting.predict(X_test)
y_predicted_train = gradientBoosting.predict(X_train)

rms_test = mean_squared_error(y_test, y_predicted_test, squared=False)
rms_train = mean_squared_error(y_train, y_predicted_train, squared=False)
print(f'train: {rms_train}\ntest: {rms_test}')
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import cross_validate as CV

scores = CV(gradientBoosting, X_train, y_train, cv=10, scoring='neg_mean_squared_error')
np.sqrt(scores['test_score']*-1)
```
